```python
#constructing a synthetic data
from sklearn.datasets import make_moons, make_circles
from matplotlib import pyplot as plt
X, y = make_moons(n_samples=1000, noise=0.1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0], X[:,1], edgecolor='k')
```

```python
#It is observative that there areintuitive clusters in the dataset.
#If we try to locate the clusters using K-Means it might not be achievable
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
y = kmeans.fit_predict(X)
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
#Detecting the clusters with DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.1, min_samples=2)
y = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
#DBSCAN detected some points as outliers! This wasn't seen by K-Means.
```
